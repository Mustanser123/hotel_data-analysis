## Data Transformation 

```python
df_agg_bookings.head()
```

```python
new_col = df_agg_bookings.apply(lambda row: row['successful_bookings']/row['capacity'], axis=1)
df_agg_bookings = df_agg_bookings.assign(occ_pct=new_col.values)
df_agg_bookings.head(3)
```

```python
# convert the occ_pct into percentage
df_agg_bookings['occ_pct'] = df_agg_bookings['occ_pct'].apply(lambda x: round(x*100, 2))
df_agg_bookings.head(3)
```

```python
df_bookings.head()
```

## Insight Generation

#### 1. What is an average occupancy rate in each of the room categories?

```python
df_agg_bookings.head(3)
```

```python
df_agg_bookings.groupby("room_category")["occ_pct"].mean()
```

print the occupancy regarding regarding the categories such as premium or elite etc so for this we have to merge the two table because both the information is not avialable in the same table

```python
df = pd.merge(df_agg_bookings, df_rooms, left_on="room_category", right_on="room_id")
df.head(4)
```

```python
# we need to drop the room_id because no need of the room_id as we have room_class
df.drop("room_id",axis=1, inplace=True)
df.head(4)
```

```python
df.groupby("room_class")["occ_pct"].mean()
```

```python
# print the occupancy only for standard 
df[df.room_class=="Standard"].occ_pct.mean()
```

#### 2. Print average occupancy rate per city

```python
df_hotels.head(3)
```

```python
df = pd.merge(df, df_hotels, on="property_id")
df.head(3)
```

```python
df.groupby("city")["occ_pct"].mean()
```

#### when was the occupancy better on weekday or weekend?

```python
df_date.head(3)
```

```python
df = pd.merge(df, df_date, left_on="check_in_date", right_on="date")
df.head(3)
```

```python
df.groupby("day_type")["occ_pct"].mean().round(2)
#so on weekend the occupancy is greater because people mostly make visits on weekend
```

#### what was the occupancy rate for different cities in the month of june

```python
df_june = df[df["mmm yy"]=="Jun 22"]
df_june.head(4)
```

```python
df_june.groupby('city')['occ_pct'].mean().round(2).sort_values(ascending=False)
```

```python
df_june.groupby('city')['occ_pct'].mean().round(2).sort_values(ascending=False).plot(kind="bar")
```
